```python
"""
Voorbeelden gebruik van CBS Open Data v3 in Python
https://www.cbs.nl/nl-nl/onze-diensten/open-data
Auteur: Jolien Oomens
Centraal Bureau voor de Statistiek
In dit voorbeeld worden gemeentegrenzen gekoppeld aan geboortecijfers om een 
thematische kaart te maken.
"""

import pandas as pd
import numpy as np
import geopandas as gpd
import cbsodata
import matplotlib.pyplot as plt

from deep_translator import GoogleTranslator

gt = GoogleTranslator(source='nl', target='en')
```

```python
# specify the custom font to use
plt.rcParams['font.family'] = 'sans-serif'
# installed font from attached 'reynold.zip'
plt.rcParams['font.sans-serif'] = 'helvetica'
```

```python
def translate(d,i=0, verbose=False):
    """
    This function translates a Dutch string into English using Google Translate.
    
    It handles multiline input, and if the string is >4000 characters long, splits the
    string, translates the parts and returns the concatenated whole English string.
    
    There is a possibility of tracking the process through the index i.
    
    Prints state messages is verbose=True.
    
    Returns None on failure.
    """
    try:
        if verbose:
            print(f"Translating {i}")
        d = d.replace('\n','# ')
        if len(d)>4000:
            if verbose:
                print("Overlong description, splitting translation")
            res = []
            for i in range(len(d)//4000):
                res.append(gt.translate(d[(i*4000):((i+1)*4000)]))
            res.append(gt.translate(d[((i+1)*4000):len(d)]))
        else:
            res = [gt.translate(d)]
        return " ".join(res).replace('# ','\n')
    except:
        return None
```

```python
# Translating the downloaded table into English (~1-2 minutes)
print("Translating titles...")
metadata['TitleEn'] = metadata.apply(lambda row: translate(row['Title'],row.name,False),axis=1)
print("Done.")

print("Translating descriptions...")
metadata['DescriptionEn'] = metadata.apply(lambda row: translate(row['Description'],row.name,False),axis=1)
print("Done.")
```

```python
plt.rcParams['legend.title_fontsize'] = 20

# Try with other measures

identifier = "83765NED"
var_plot= "GemiddeldeWoningwaarde_35"
resolution=7

# Download birth rates and remove spaces from region codes
data = pd.DataFrame(cbsodata.get_data(identifier, select = ['WijkenEnBuurten', 'Codering_3', var_plot]))
data['Codering_3'] = data['Codering_3'].str.strip()

# Retrieve the municipal boundaries map from PDOK
geodata_url = 'https://geodata.nationaalgeoregister.nl/cbsgebiedsindelingen/wfs?request=GetFeature&service=WFS&version=2.0.0&typeName=cbs_gemeente_2017_gegeneraliseerd&outputFormat=json'
gemeentegrenzen = gpd.read_file(geodata_url)

# Link Statistics Netherlands data to geodata with region codes
gemeentegrenzen = pd.merge(gemeentegrenzen, data,
                           left_on = "statcode", right_on = "Codering_3")

gemeentegrenzen_c=gemeentegrenzen.to_crs('epsg:4326').h3.polyfill(resolution, explode=True)
gemeentegrenzen_c=gemeentegrenzen_c.dropna(subset=['h3_polyfill'])
gdf_h3 = gemeentegrenzen_c.set_index('h3_polyfill').h3.h3_to_geo_boundary()

ax = gdf_h3.plot(figsize=(15, 15), column=var_plot, cmap='viridis_r',legend=True, legend_kwds={'label':'Average House Value','orientation': "horizontal",'shrink': 0.5})

#ax.set_title('Average House Value',font='Arial',fontsize=25)
ax.axis('off')

fig = ax.figure
cb_ax = fig.axes[1] 
cb_ax.tick_params(labelsize=14)

#leg = ax.legend()
#leg.set_title('Average House Value',prop={'size':14})

plt.savefig('challenge_hex1.jpg')
```

```python
# Try with other measures

identifier = "83765NED"
var_plot= "MateVanStedelijkheid_104"
resolution=7
color_map='magma_r'

# Download birth rates and remove spaces from region codes
data = pd.DataFrame(cbsodata.get_data(identifier, select = ['WijkenEnBuurten', 'Codering_3', var_plot]))
data['Codering_3'] = data['Codering_3'].str.strip()

# Retrieve the municipal boundaries map from PDOK
geodata_url = 'https://geodata.nationaalgeoregister.nl/cbsgebiedsindelingen/wfs?request=GetFeature&service=WFS&version=2.0.0&typeName=cbs_gemeente_2017_gegeneraliseerd&outputFormat=json'
gemeentegrenzen = gpd.read_file(geodata_url)

# Link Statistics Netherlands data to geodata with region codes
gemeentegrenzen = pd.merge(gemeentegrenzen, data,
                           left_on = "statcode", right_on = "Codering_3")

gemeentegrenzen_c=gemeentegrenzen.to_crs('epsg:4326').h3.polyfill(resolution, explode=True)
gemeentegrenzen_c=gemeentegrenzen_c.dropna(subset=['h3_polyfill'])
gdf_h3 = gemeentegrenzen_c.set_index('h3_polyfill').h3.h3_to_geo_boundary()

gdf_h3=gdf_h3.to_crs('epsg:28992')
gdf_h3[var_plot]=gdf_h3[var_plot]*(-1)
ax = gdf_h3.plot(figsize=(15, 15), column=var_plot, cmap=color_map,legend=True, legend_kwds={'orientation': "horizontal",'shrink': 0.5,'ticks': [-1,-5]})

ax.axis('off')

fig = ax.figure
cb_ax = fig.axes[1] 
cb_ax.tick_params(labelsize=14,pad=0.2)

ax.set_title('Degree of urbanization',font='Arial',fontsize=21, y=-0.2)

#ax.set_title(loc='bottom')
ax.axis('off')

plt.savefig('challenge_hex4.jpg',bbox_inches='tight')
```
